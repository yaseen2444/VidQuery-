```python
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled
video_id = "Gfr50f6ZBvo" # only the ID, not full URL
try:
    # If you dont care which language, this returns the best one
    transcript_list = YouTubeTranscriptApi.get_transcript(video_id, languages=["en"])

    # Flatten it to plain text
    transcript = " ".join(chunk["text"] for chunk in transcript_list)
    print(transcript)

except TranscriptsDisabled:
    print("No captions available for this video.")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5")

# Step 2: Split your transcript into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.create_documents([transcript])  # transcript should be a string

# Step 3: Create FAISS vector store with the new embedding model
vector_store = FAISS.from_documents(chunks, embeddings)
print(vector_store)
```

```python
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 4})
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "mistralai/Mistral-7B-Instruct-v0.3"  # your local path or model hub name
hf_token = "hf_foIAvFvZwNPiWHqlskVRuilCxWSobIWyX"  # or set it as an env var

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    load_in_8bit=True,
    use_auth_token=hf_token  # optional if using private models
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

text_gen_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    max_new_tokens=512,
    temperature=0.2,
    return_full_text=False,
)

llm = HuggingFacePipeline(pipeline=text_gen_pipeline)
```

```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(
    template="""
You are a helpful assistant.
Answer ONLY from the provided transcript context.
If the context is insufficient, just say you don't know.

{context}
Question: {question}
""",
    input_variables=["context", "question"]
)
```

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 4}),
    chain_type="stuff",  # or "refine" or "map_reduce" if you need
    chain_type_kwargs={"prompt": prompt}
)
```

```python
query = "What is the speaker explaining about AI models?"
response = qa_chain.run(query)
print(response)
```
